```python
def transform_gender(df):
    df["Sex"] = df["Sex"].map({"female": 0, "male": 1})
    return df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def label_encode_column(df, column_name):
    df[column_name] = LabelEncoder().fit_transform(df[column_name] )
    return df
```

```python
def drop_column(df, column_name):
    df = df.drop(column_name, axis=1)
    return df
```

```python
# Versiunea 2 - pipeline
train_df.pipe(label_encode_column, "Sex").pipe(label_encode_column, "Embarked").pipe(drop_column, "Name").pipe(drop_column, "Ticket")
```
