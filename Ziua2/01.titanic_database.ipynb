```python
df["Gender"] = df["Sex"] == "male"
df.head(2)
```

```python
df["Gender"] = df["Sex"].map({"female": 0, "male": 1})
df.head()
```

```python
df = df.drop("Sex", axis=1)
df.head()
```

```python
df["Name"].nunique()
```

```python
for key in ('Pclass', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'):
    print("key: ", key, df[key].nunique())
```

```python
df["Embarked"] = df["Embarked"].map({'S':3, 'C':1, 'Q':2, 'nan':0})
df.head()
```

```python
df.dtypes
```

```python
df["Cabin"].unique(), df["Cabin"].nunique()
```

```python
set([str(x)[0] for x in  df["Cabin"].unique()])
```

```python
df["Cabin_LETTER"] = df["Cabin"].astype(np.str_).str[0]
df.head()
```

```python
df["Cabin_LETTER"] = df["Cabin_LETTER"].map({'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'T':8, 'n':0})
df
```
